```python
import json
import boto3

# Initialize the Bedrock runtime client
bedrock_runtime = boto3.client('bedrock-runtime', region_name='us-east-1')

def call_openai_gpt(prompt):
    prompt_config = {
        "prompt": prompt,
        "max_gen_len": 512,
        "temperature": 0.5,
        "top_p": 0.9
    }

    body = json.dumps(prompt_config)

    modelId = "meta.llama3-70b-instruct-v1:0"
    accept = "application/json"
    contentType = "application/json"

    try:
        response = bedrock_runtime.invoke_model(
            body=body, modelId=modelId, accept=accept, contentType=contentType
        )
        response_body = json.loads(response.get("body").read())
        
        # Extracting the generated text from the response
        model_results = response_body.get("generation")
        
        if model_results is None:
            raise ValueError("Generated text not found in the response")
        
        return model_results
    except bedrock_runtime.exceptions.AccessDeniedException as e:
        raise Exception(f"Access Denied: {e}")
    except Exception as e:
        raise Exception(f"An error occurred: {e}")


```

```python
def get_specific_tables_metadata(connection, database_name, schema_name, tables_df, num_samples = 2):
    metadata = ""
    working_tables = tables_df[tables_df['Flag'] == 1]
    
    for _, table_row in working_tables.iterrows():
        table_name = table_row['Table_name']
        metadata += f"\nMetadata for table {database_name}.{schema_name}.{table_name}:\n"
        sql = f"SHOW COLUMNS FROM {schema_name}.{table_name}"
        # sql = f"SHOW COLUMNS FROM TABLE {database_name}.{schema_name}.{table_name}"
        
        result = connection(sql)
        # result = cursor.fetchall()
        create_table_statement = f"CREATE TABLE {schema_name}.{table_name} ("
        for row in result:
            column_name = row[0]
            data_type = row[1]
            create_table_statement += f"\n {column_name} {data_type},"
        create_table_statement = create_table_statement.rstrip(',') + "\n);"
        metadata += create_table_statement + "\n"

        sample_sql = f"SELECT * FROM {schema_name}.{table_name} LIMIT {num_samples};"
        sample_rows = connection(sample_sql)


        metadata += f"\nSample rows from {table_name} table:\n"
        for row in sample_rows:
            metadata += str(row) + "\n"
    # cursor.close()
    return metadata


df = pd.read_csv('health_data_schema_1.csv')
database_name = 'healthcare'
schema_name = 'healthcare'
table_info = get_specific_tables_metadata(connection, database_name, schema_name, df)
```

```python
import openai
import tiktoken

openai.api_key = 'your api key'

def call_openai_gpt(prompt):
    # Initialize the tokenizer for the specified model
    encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")
    
    # Token count of the input prompt
    input_tokens = encoding.encode(prompt)
    input_token_count = len(input_tokens)
    
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # Specify the model
        messages=[
            {"role": "user", "content": prompt}
        ],
        max_tokens=512,
        temperature=0.5,
        top_p=0.9
    )
    
    # Extracting the generated text from the response
    model_results = response.choices[0].message['content'].strip()
    
    # Token count of the output response
    output_tokens = encoding.encode(model_results)
    output_token_count = len(output_tokens)
    
    return model_results, input_token_count, output_token_count


```

```python
prompt = 'Give 5 foreign cat breeds'
result, input_tokens, output_tokens = call_openai_gpt(prompt)
print(f"Result: {result}")
print(f"Input token count: {input_tokens}")
print(f"Output token count: {output_tokens}")
```
